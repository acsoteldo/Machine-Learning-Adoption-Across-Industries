```python
import pandas as pd
data = pd.read_csv('Tech_trend.csv')
```

```python
data.head()
data.info()
data.describe()
```

```python
data.dropna()
data.drop_duplicates()
```

```python
data.drop(columns=['Timestamp_in_seconds'], inplace=True)
```

```python
data.fillna('unknown', inplace=True)  # Replace NaN with 'unknown'
```

```python
data['Q1'] = data['Q1'].str.strip()
data['Q2'] = data['Q2'].str.strip()
data['Q3'] = data['Q3'].str.strip()
data['Q4'] = data['Q4'].str.strip()
data['Q5'] = data['Q5'].str.strip()
data['Q6'] = data['Q6'].str.strip()
data['Q7'] = data['Q7'].str.strip()
data['Q8'] = data['Q8'].str.strip()
data['Q9'] = data['Q9'].str.strip()
data['Q10'] = data['Q10'].str.strip()
data['Q11'] = data['Q11'].str.strip()
data['Q12'] = data['Q12'].str.strip()
data['Q13'] = data['Q13'].str.strip()
data['Q14'] = data['Q14'].str.strip()
data['Q15'] = data['Q15'].str.strip()
data['Q16'] = data['Q16'].str.strip()
data['Q17'] = data['Q17'].str.strip()
data['Q18'] = data['Q18'].str.strip()
data['Q19'] = data['Q19'].str.strip()
data['Q20'] = data['Q20'].str.strip()
```

```python
data.rename(columns={'Unnamed: 0': 'Person_ID'}, inplace=True)
data.rename(columns={'Q1': 'Age'}, inplace=True)
data.rename(columns={'Q2': 'Gender'}, inplace=True)
data.rename(columns={'Q3': 'Country_of_Residence'}, inplace=True)
data.rename(columns={'Q4': 'Highest_Level_of_Formal_Education'}, inplace=True)
data.rename(columns={'Q5': 'Current_Role'}, inplace=True)
data.rename(columns={'Q6': 'Yrs_of_Programming_Experience'}, inplace=True)
data.rename(columns={'Q7': 'Primary_Language_Recommendation_for_Aspiring_Data_Scientists'}, inplace=True)
data.rename(columns={'Q8': 'Preferred_Computing_Platform_for_Data_Science_Projects'}, inplace=True)
data.rename(columns={'Q9': 'Total_TPU_Usage_Count'}, inplace=True)
data.rename(columns={'Q10': 'Yrs_of_ML_Experience'}, inplace=True)
data.rename(columns={'Q11': 'Industry_of_Current_Employer_or_Contract'}, inplace=True)
data.rename(columns={'Q12': 'Company_Size'}, inplace=True)
data.rename(columns={'Q13': 'Number_of_Data_Science_Workload_Handlers'}, inplace=True)
data.rename(columns={'Q14': 'Employers_Adoption_of_ML'}, inplace=True)
data.rename(columns={'Q15': 'Annual_Compensation_USD'}, inplace=True)
data.rename(columns={'Q16': 'Total_Expenditure_on_ML_and_Cloud_Services_Past_5_Years_USD'}, inplace=True)
data.rename(columns={'Q17': 'Best_Developer_Experience_among_Cloud_Platforms'}, inplace=True)
data.rename(columns={'Q18': 'Most_Used_Big_Data_Product'}, inplace=True)
data.rename(columns={'Q19': 'Primary_Business_Intelligence_Tool'}, inplace=True)
data.rename(columns={'Q20': 'Primary_Data_Analysis_Tool_Used_at_Work_or_School'}, inplace=True)
```

```python
data['Number_of_Data_Science_Workload_Handlers'] = data['Number_of_Data_Science_Workload_Handlers'].astype(str)

# Perform the replacements
data['Number_of_Data_Science_Workload_Handlers'] = data['Number_of_Data_Science_Workload_Handlers'].replace({
    '2023-02-01 00:00:00': '1-2',
    '2023-04-03 00:00:00': '3-4',
    '2023-09-05 00:00:00': '5-9',
    '2014-10-01 00:00:00': '10-14'
})
```

```python
mapping = {
    '1-2 years': 2,
    '10-20 years': 15,
    '2-3 years': 3,
    '20 or more years': 20,
    '3-4 years': 4,
    '4-5 years': 5,
    '5-10 years': 10,
    'Under 1 year': 1,
    'I do not use machine learning methods': 0
}

# Create a new column with integers representing years of ML experience
data['Yrs_of_ML_Experience_Int'] = data['Yrs_of_ML_Experience'].map(mapping)
```

```python
mapping = {
    '1-3 years': 2,
    '10-20 years': 15,
    '20+ years': 20,
    '3-5 years': 4,
    '5-10 years': 10,
    '< 1 years': 1,
    'I have never written code': 0
}

# Create a new column with integers representing years of ML experience
data['Yrs_of_Programming_Experience_Int'] = data['Yrs_of_Programming_Experience'].map(mapping)
```

```python
mapping = {
    '0-49 employees': 1,
    '50-249 employees': 2,
    '250-999 employees': 3,
    '1000-9,999 employees': 4,
    '10,000 or more employees': 5
}

data['Company_Size_Rank'] = data['Company_Size'].map(mapping)
```
